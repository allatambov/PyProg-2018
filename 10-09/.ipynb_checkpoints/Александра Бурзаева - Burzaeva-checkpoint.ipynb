## Отступы, пробелы, табуляция 

В версиях Python выше **3.0** нельзя смешивать использование пробелов и табуляции в отступах.
**Пробелы** считаются предпочтительным методом для отступов. 

На каждый уровень отступа необходимо использовать _4 пробела_.

На одной строчке нельзя расположить более **79 символов**. 

Для логического разделения материала советуется использовать пустые строки в функциях.

## Импорты

Каждый импорт должен быть расположен на новой строке.
_Пример правильного расположения импортов ниже_.



```python
import math # импортируем модуль math

import math as ma 

```

Импорты всегда помещается в **верхнюю часть** файла, сразу после комментариев к модулю и доклингов, а также перед константами модуля.

Импорты должны быть сгруппированы в следующем порядке:

1. Импорты стандартных библиотек.
2. Импорты сторонних бибилиотек.
3. Локальный импорт приложений / библиотек.

Между каждой группой импортов необходимо вставлять _пустую строку_.

## Комментарии и стили

Использовать комментарии, которые находятся на одной и той же строке с кодом, лучше всего только в тех обстоятельствах, когда **без них нельзя понять задумку автора кода**.

Существует много разных стилей. Обычно различают _следующие_:

* b (одиночная маленькая буква)
* B (одиночная заглавная буква)
* lowercase (слово в нижнем регистре)
* lower_case_with_underscores (слова из маленьких букв с подчеркиваниями)
* UPPERCASE (заглавные буквы)
* UPPERCASE_WITH_UNDERSCORES (слова из заглавных букв с подчеркиваниями)
* CapitalizedWords (слова с заглавными буквами, или CapWords, или CamelCase)
* mixedCase (отличается от CapitalizedWords тем, что первое слово начинается с маленькой буквы)
* Capitalized_Words_With_Underscores (**!некрасиво!**)
